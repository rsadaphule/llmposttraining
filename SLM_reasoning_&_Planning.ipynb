```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

print("Running on GPU:", torch.cuda.is_available())
```

```python
model_name = "Qwen/Qwen2.5-0.5B-Instruct"  # small 135M param model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name).to("cuda")
```

```python
prompt = """
### Instruction:
If Sarah has 24 apples and eats 3 each day,
how many days until she has fewer than 5 left?
Show your steps clearly.

### Response:
"""
inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
outputs = model.generate(**inputs, max_new_tokens=400)
print(tokenizer.decode(outputs[0]))
```

```python
prompt = """
### Instruction:
Summarize the following passage as if you were:
1. A high school teacher (≤ 2 sentences).
2. A financial analyst (≤ 2 sentences).
Present both summaries separately.
### Input:
Climate policy implementation creates significant economic disruption and opportunity across multiple sectors,
requiring careful analysis of both immediate costs and long-term financial benefits.
key metrics include the estimated $4-5 trillion annual global investment needed for net-zero transition,
carbon pricing mechanisms that currently range from $1-130 per ton of CO2 across different jurisdictions, and sectoral
shifts where renewable energy costs have dropped 70-90% over the past decade while fossil fuel companies face stranded asset risks.
Traditional industries face compliance costs averaging 2-8% of revenue depending on sector, while clean technology markets are projected to
reach $2.5 trillion by 2030, creating investment opportunities in areas like battery storage, green hydrogen, and carbon capture technologies.
climate policies are driving curriculum transformation and skills retraining at unprecedented scales.
High schools are integrating STEM education focused on sustainability challenges, with states like California requiring climate
science education and others adding environmental literacy standards to graduation requirements. The U.S. Bureau of Labor Statistics
 projects that 8 of the 10 fastest-growing occupations through 2031 will be in renewable energy and environmental services, requiring educational
 institutions to develop new technical programs while also preparing students for traditional careers that are being transformed by sustainability
 requirements—from sustainable agriculture to green finance.
The social dimensions present both challenges and opportunities that educators and analysts must understand.
 Climate policies can exacerbate existing inequalities if not carefully designed, as energy costs may rise disproportionately for
 low-income households, while wealthy communities benefit first from electric vehicle incentives and solar installations.
 However, targeted policies like weatherization programs, community solar projects, and green job training in disadvantaged areas
 can create pathways for economic mobility. Teachers need frameworks for discussing these trade-offs with students, while financial
 analysts must account for social license and community acceptance in their risk assessments, as projects without local support face higher
 failure rates and regulatory challenges.
### Response:
"""
inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
outputs = model.generate(**inputs, max_new_tokens=400)
print(tokenizer.decode(outputs[0]))
```

```python
from datasets import load_dataset

dataset = load_dataset("yahma/alpaca-cleaned")
print(dataset)
```

```python
dataset["train"][0]
```

```python
def format_example(example):
    instr = example["instruction"]
    inp = example.get("input", "")
    out = example["output"]
    if inp and len(inp.strip()) > 0:
        # If there's additional input, include it
        prompt_text = f"### Instruction:\n{instr}\n\n### Input:\n{inp}\n\n### Response:\n"
    else:
        prompt_text = f"### Instruction:\n{instr}\n\n### Response:\n"
    # We want the model to generate the output following the prompt
    full_text = prompt_text + out
    return {"text": full_text}

# Apply formatting
formatted_dataset = dataset["train"].map(format_example)
```

```python
formatted_dataset[0]
```

```python
tokenizer.pad_token = tokenizer.eos_token  # use <|endoftext|> as padding token

def tokenize(batch):
    return tokenizer(batch["text"], padding="max_length", truncation=True, max_length=512)

tokenized_dataset = formatted_dataset.map(tokenize, batched=True, remove_columns=["text"])
# Split into train/validation
tokenized_dataset = tokenized_dataset.train_test_split(test_size=0.05, seed=42)
train_data = tokenized_dataset["train"]
val_data = tokenized_dataset["test"]
```

```python
from transformers import Trainer, TrainingArguments, DataCollatorForLanguageModeling

# Data collator for causal LM – this will handle creating batches and masks
data_collator = DataCollatorForLanguageModeling(tokenizer, mlm=False)  # mlm=False means this is for causal LM

training_args = TrainingArguments(
    output_dir="qwen-slm-finetune-checkpoints",
    eval_strategy="steps",
    eval_steps=200,
    save_steps=200,
    logging_steps=50,
    num_train_epochs=1,            # We'll do 1 epoch for demonstration; can be increased
    per_device_train_batch_size=8, # adjust based on GPU memory; 8 is okay for 135M on a T4
    per_device_eval_batch_size=8,
    gradient_accumulation_steps=4, # accumulate gradients to simulate batch size of 32
    learning_rate=2e-5,            # a common learning rate for fine-tuning
    fp16=True,                     # use mixed precision for speed
    save_total_limit=2,            # keep only 2 checkpoints to save space
)
```

```python
trainer = Trainer(
    model=model,
    tokenizer=tokenizer,
    args=training_args,
    data_collator=data_collator,
    train_dataset=train_data,
    eval_dataset=val_data
)

trainer.train()
```

```python
finetuned_model = trainer.model.to("cuda")
```

```python
finetuned_model.save_pretrained("qwen-finetuned-600m")
tokenizer.save_pretrained("qwen-finetuned-600m")
```

```python

MODEL_DIR = "qwen-finetuned-600m"   # same folder you saved above

# Load tokenizer & model
tokenizer = AutoTokenizer.from_pretrained(MODEL_DIR)
if tokenizer.pad_token is None:   # many GPT-like SLMs don’t have a pad token
    tokenizer.pad_token = tokenizer.eos_token

model_postft = AutoModelForCausalLM.from_pretrained(MODEL_DIR)
model_postft.to("cuda" if torch.cuda.is_available() else "cpu")
model_postft.eval()
```

```python
prompt = """### Instruction:
If Sarah has 24 apples and eats 3 each day,
how many days until she has fewer than 5 left?
Show your steps clearly.
### Response:
"""
inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
outputs = model_postft.generate(**inputs, max_new_tokens=300)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
